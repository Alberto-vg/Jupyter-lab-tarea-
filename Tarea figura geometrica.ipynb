```python
import numpy as np
import matplotlib.pyplot as plt

N_PUNTAS = 8
t = np.arange(-6.2, 6.2, 0.01)

factor = np.round(np.sqrt(np.cos(np.cos(8.2*t))))
x = 10*np.sin(2.78*t) * factor
y = 9*(np.cos(2.78*t))**2 * np.sin(np.sin(8.2*t))

# Ahora puntos como (N, 2)
pts = np.array([x, y]).T          # shape: (N, 2)
N = pts.shape[0]

for k in range(N_PUNTAS):
    theta = k * (2*np.pi / N_PUNTAS)
    R = np.array([[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta),  np.cos(theta)]])

    # Multiplicación 
    rot = np.zeros_like(pts)      # (N, 2)
    for j in range(N):
        rot[j, 0] = R[0, 0]*pts[j, 0] + R[0, 1]*pts[j, 1]
        rot[j, 1] = R[1, 0]*pts[j, 0] + R[1, 1]*pts[j, 1]

    plt.plot(rot[:, 0], rot[:, 1], linewidth=1)

plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
