```python
# Curva paramtrica tipo "roseta" 
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la frmula
a, b, c = 1, 49, 0

# Tiempo (suficiente para cerrar la curva porque a y b son enteros)
t = np.arange(0.0, 2*np.pi, 0.001)   

# x(t), y(t) segn la imagen:
# x(t) = cos(a t) + cos(b t)/2 + sin(c t)/3
# y(t) = sin(a t) + sin(b t)/2 + cos(c t)/3
x = np.cos(a*t) + np.cos(b*t)/2.0 + np.sin(c*t)/3.0
y = np.sin(a*t) + np.sin(b*t)/2.0 + np.cos(c*t)/3.0

# Apilo en un arreglo 2xN 
pts = np.array([x, y])               # (2, N)
N = pts.shape[1]

# Trazo de la curva por segmentos 
for j in range(N - 1):
    plt.plot([pts[0, j], pts[0, j+1]],
             [pts[1, j], pts[1, j+1]],
             linewidth=1)

# Ajuste para no deformar
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
